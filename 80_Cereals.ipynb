```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Reading the dataset
df=pd.read_csv("/content/sample_data/cereal.csv")
#Display column names
print(df.columns)
```

```python
#Display Top 5 Values
print(df.head())
```

```python
#Display bottom 5 values
print(df.tail())
```

```python
#Gives the descriptive statistics of the dataset
print(df.describe().to_string())
```

```python
#Gives the concise information of the dataframe
print(df.info())
```

```python
#Displays the number of rows and columns
print(df.shape)
```

```python
#Displays the number of duplicated rows
print(df.duplicated().sum())
```

```python
#Barchart
sns.barplot(x="calories",y="fiber",data=df)
plt.show()
```

```python
#Barchart
sns.catplot(x="type",y="protein",data=df,kind="bar")
plt.show()
```

```python
#Barchart for sugars
plt.figure(figsize=(15,5))
sns.barplot(x=df["name"],y=df["sugars"])
plt.xticks(rotation=90)
plt.show()
```

```python
#Crosstab
pd.crosstab(df["type"],df["protein"])
```

```python
#Visualization of above crosstab
pd.crosstab(df["type"],df["protein"]).plot(kind="bar")
plt.xticks(rotation=0)
plt.show()
```

```python
#Barchart for carbohydrate
plt.figure(figsize=(16,5))
sns.barplot(x=df["name"],y=df["carbo"])
plt.xticks(rotation=90)
plt.show()
```

```python
#Barchart for vitamins
plt.figure(figsize=(16,5))
sns.barplot(x=df["name"],y=df["vitamins"])
plt.xticks(rotation=90)
plt.show()
```

```python
#Barchart for fiber
plt.figure(figsize=(16,5))
sns.barplot(x=df["name"],y=df["fiber"])
plt.xticks(rotation=90)
plt.show()
```
